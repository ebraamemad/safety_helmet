```python

import optuna
import mlflow
from ultralytics import YOLO
import os


def objective(trial):
    lr0 = trial.suggest_float("lr0", 0.01, 0.1, log=True)
    dropout = trial.suggest_float("dropout", 0.0, 0.5)
    batch = trial.suggest_categorical("batch", [32, 64])

    
    with mlflow.start_run(nested=True):
        mlflow.log_params({
            "lr0": lr0,
            "dropout": dropout,
            "batch": batch
        })

        
        results = model.train(
            data=r"E:\projects of camp\safety-helmet\data\data.yaml",
            epochs=3,
            imgsz=128,
            batch=batch,
            lr0=lr0,
            dropout=dropout,
            name="optuna_run",
            project="yolo_optuna_mlflow"
        )

        # Read last mAP50 from results.csv
        # مهمة في مقارنة النتائج بين التجارب المختلفة
        try:
            import pandas as pd
            df = pd.read_csv(f"yolo_optuna_mlflow/optuna_run/results.csv")
            mAP = df["metrics/mAP_0.5"].iloc[-1]
            mlflow.log_metric("mAP_0.5", mAP)
            return mAP
        except:
            return 0.0

study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=5)
```
